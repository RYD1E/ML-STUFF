# Lets start

Data visualization is a powerful tool for understanding and communicating insights from datasets. Choosing the right type of graph can significantly enhance the effectiveness of your visualizations. Here are some considerations for when to choose a graph to visualize data in Python:


## 1. **Comparing Values**

When you want to compare individual values or sets of values, consider the following graphs:

- **Bar Chart**: Suitable for comparing discrete categories. It's great for visualizing categorical data or comparing values across different groups.

- **Column Chart**: Similar to a bar chart, but with vertical columns. It's also used for comparing discrete categories.

- **Line Chart**: Useful for showing trends and changes over time. It's ideal for visualizing continuous data.


Certainly! Let's go through each of these types of graphs with examples and analysis:

#### a. Bar Chart

**Example:**

Suppose you have a dataset of monthly sales figures for different products in a store. You want to compare the sales performance of each product.

```python
import matplotlib.pyplot as plt

products = ['Product A', 'Product B', 'Product C', 'Product D']
sales = [12000, 15000, 9000, 11000]

plt.bar(products, sales)
plt.xlabel('Products')
plt.ylabel('Sales (in dollars)')
plt.title('Monthly Sales Comparison')
plt.show()
```

**Explaination:**

A bar chart is a visual representation of categorical data where the categories are displayed as rectangular bars with lengths proportional to the values they represent. It's used to compare discrete categories. Each bar represents a category, and the height of the bar corresponds to the value it represents. Bar charts are typically used for visualizing categorical data or comparing values across different groups.

#### b. Column Chart

**Example:**

Let's use the same dataset as before, but this time, we'll use a column chart.

```python
plt.barh(products, sales)
plt.xlabel('Sales (in dollars)')
plt.ylabel('Products')
plt.title('Monthly Sales Comparison')
plt.show()
```

**Explaination:**

A column chart is similar to a bar chart, but with vertical columns instead of horizontal bars. It's used for the same purpose as a bar chart, to compare discrete categories. The choice between a bar chart and a column chart depends on the orientation that best suits the data and the presentation.

#### c. Line Chart

**Example:**

Suppose you have a dataset of monthly temperature records for a year. You want to visualize the temperature trend over time.

```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
temperatures = [5, 7, 11, 16, 21, 26, 30, 29, 24, 18, 12, 7]

plt.plot(months, temperatures, marker='o')
plt.xlabel('Months')
plt.ylabel('Temperature (Â°C)')
plt.title('Monthly Temperature Trend')
plt.grid(True)
plt.show()
```

**Explaination:**

A line chart is a graphical representation of data in which values are plotted as points and connected by lines. It's particularly effective for showing trends and changes over time. The x-axis usually represents time or a continuous variable, while the y-axis represents the corresponding values. Line charts are ideal for visualizing continuous data.

### Conclusion:

- **Bar Chart** and **Column Chart** are effective for comparing discrete categories. They are particularly useful when you have categorical data or want to compare values across different groups.

- **Line Chart** is ideal for showing trends and changes over time. It is well-suited for visualizing continuous data, making it a powerful tool for time series analysis.


````python
import matplotlib.pyplot as plt

# Data for the bar chart
products = ['Product A', 'Product B', 'Product C', 'Product D']
sales = [12000, 15000, 9000, 11000]

# Data for the line chart
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
temperatures = [5, 7, 11, 16, 21, 26, 30, 29, 24, 18, 12, 7]

# Create subplots
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

# Plot the bar chart
axs[0].bar(products, sales)
axs[0].set_xlabel('Products')
axs[0].set_ylabel('Sales (in dollars)')
axs[0].set_title('Monthly Sales Comparison')

# Plot the horizontal bar chart
axs[1].barh(products, sales)
axs[1].set_xlabel('Sales (in dollars)')
axs[1].set_ylabel('Products')
axs[1].set_title('Monthly Sales Comparison')

# Plot the line chart
axs[2].plot(months, temperatures, marker='o')
axs[2].set_xlabel('Months')
axs[2].set_ylabel('Temperature (Â°C)')
axs[2].set_title('Monthly Temperature Trend')
axs[2].grid(True)

# Adjust layout to prevent overlapping
plt.tight_layout()

# Show the plots
plt.show()
````

## 2. **Displaying Distributions**

When you need to understand the distribution of a dataset, consider these graphs:

- **Histogram**: Shows the frequency distribution of a single variable. It's useful for understanding the underlying distribution of the data.

- **Box Plot (Box-and-Whisker Plot)**: Provides a summary of the distribution, showing the median, quartiles, and potential outliers.



Certainly! Let's explore the graphs for displaying distributions along with examples and analysis:

### Histogram

**Example:**

Suppose you have a dataset representing the ages of a group of people. You want to understand the distribution of ages.

```python
import matplotlib.pyplot as plt

ages = [22, 25, 28, 30, 32, 35, 38, 40, 45, 50, 55, 60, 65, 70]

plt.hist(ages, bins=5, edgecolor='black')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution')
plt.show()
```

**Explanation:**

- Histograms are used to represent the distribution of a single variable (in this case, ages).
- The x-axis represents the range of values (in this case, different age intervals).
- The y-axis represents the frequency or count of how many data points fall within each interval.
- The bins parameter determines the number of intervals or bars. In this example, it's set to 5, meaning there will be 5 age intervals.
- The edgecolor parameter sets the color of the edges of the bars.

### Box Plot (Box-and-Whisker Plot)

**Example:**

Let's use the same dataset of ages, but this time, we'll create a box plot.

```python
plt.boxplot(ages, vert=False)
plt.xlabel('Age')
plt.title('Age Distribution')
plt.show()
```

**Explanation:**

- Box plots (or box-and-whisker plots) provide a summary of the distribution of a dataset.
- The box in the plot represents the interquartile range (IQR), which contains the middle 50% of the data. The line inside the box represents the median.
- The "whiskers" extend to show the range of the data, excluding any potential outliers.
- Outliers are individual data points that fall significantly outside the range of the rest of the data.

### Conclusion:

- **Histogram** is an excellent choice when you want to understand the frequency distribution of a single variable. It helps in visualizing the spread and central tendency of the data.

- **Box Plot** provides a concise summary of the distribution, including the median, quartiles, and potential outliers. It's especially useful when you need a quick overview of the data's spread and central tendency.


````python
import matplotlib.pyplot as plt

ages = [22, 25, 28, 30, 32, 35, 38, 40, 45, 50, 55, 60, 65, 70]


# Create subplots
fig, axs = plt.subplots(1, 2, figsize=(15, 5))
axs[0].hist(ages, bins=5, edgecolor='black')
axs[0].set_xlabel('Age')
axs[0].set_ylabel('Frequency')
axs[0].set_title('Age Distribution')

axs[1].boxplot(ages, vert=False)
axs[1].set_xlabel('Age')
axs[1].set_title('Age Distribution')

# Adjust layout to prevent overlapping
plt.tight_layout()

# Show the plots
plt.show()
````

## 3. **Showing Relationships**

When you want to visualize relationships between variables, consider these options:

- **Scatter Plot**: Useful for visualizing the relationship between two continuous variables. It helps identify correlations or patterns.

- **Bubble Chart**: Extends a scatter plot by adding a third dimension, typically represented by the size of the markers.

- **Heatmap**: Great for displaying the correlation between multiple variables in a grid format.


Certainly! Let's explore the graphs for showing relationships along with examples and analysis:

#### a.Scatter Plot

**Example:**

Suppose you have a dataset representing the relationship between hours spent studying and exam scores for a group of students.

```python
import matplotlib.pyplot as plt

hours_studied = [3, 4, 2, 6, 8, 7, 5, 9, 1, 3]
exam_scores = [60, 70, 50, 80, 90, 85, 75, 95, 45, 65]

plt.scatter(hours_studied, exam_scores)
plt.xlabel('Hours Studied')
plt.ylabel('Exam Score')
plt.title('Relationship between Hours Studied and Exam Score')
plt.show()
```

**Explanation:**

- A scatter plot is a graph used to visualize the relationship between two continuous variables.
- It consists of individual data points, where each point represents a pair of values from the two variables.
- The x-axis typically represents one variable, while the y-axis represents the other variable.
- Scatter plots are useful for identifying correlations, patterns, and trends between the two variables.
- If the points form a recognizable pattern, such as a linear trend, it indicates a relationship between the variables.

#### b. Bubble Chart

**Example:**

Suppose you want to extend the previous example by adding a third dimension, which represents the number of assignments completed by each student.

```python
assignments_completed = [4, 3, 2, 5, 7, 6, 4, 8, 1, 3]

plt.scatter(hours_studied, exam_scores, s=assignments_completed*10)
plt.xlabel('Hours Studied')
plt.ylabel('Exam Score')
plt.title('Relationship between Hours Studied, Exam Score, and Assignments Completed')
plt.show()
```

**Explanation:**

- A bubble chart is an extension of a scatter plot that incorporates a third dimension by varying the size of markers.
- In addition to the x and y coordinates, the size of each marker is determined by a third variable.
- This third variable can represent another aspect of the data, providing a way to visualize relationships involving three variables.
- Bubble charts are effective when you want to convey information about the magnitude or importance of a third variable while showing its relationship with the other two variables.

#### c. Heatmap

**Example:**

Suppose you have a dataset containing the correlation coefficients between multiple variables.

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Generate random student scores for five subjects
np.random.seed(0)  # Set seed for reproducibility
scores = np.random.randint(50, 100, size=(30, 5))  # Generate random scores between 50 and 100

# Create a DataFrame with the data
data = {
    'Math': scores[:, 0],
    'English': scores[:, 1],
    'Science': scores[:, 2],
    'History': scores[:, 3],
    'Geography': scores[:, 4]
}

df = pd.DataFrame(data)

# Create a correlation matrix
corr_matrix = df.corr()

# Generate the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap of Subjects')
plt.show()

```

**Explanation:**

In this example, we first create a DataFrame with the student scores in different subjects.
- Next, we calculate the correlation matrix (corr_matrix) which shows how each subject correlates with every other subject. The values range from -1 (perfect negative correlation) to 1 (perfect positive correlation).
- We then use seaborn's heatmap function to visualize the correlation matrix. The annot=True parameter displays the correlation values on the heatmap.
- The cmap='coolwarm' parameter sets the color scheme of the heatmap.

### Conclusion:

- **Scatter plots** are effective for visualizing relationships between two continuous variables and identifying correlations or patterns.

- **Bubble charts** extend scatter plots by adding a third dimension, allowing for visualization of relationships involving three variables.

- **Heatmaps** are excellent for displaying the correlation between multiple variables in a grid format, making it easier to identify patterns in large datasets.

The choice between these visualization techniques depends on the nature of your data and the relationships you want to explore. Each graph serves a different purpose in visualizing and understanding relationships between variables.

````python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

# Generate random student scores for five subjects
np.random.seed(0)  # Set seed for reproducibility
scores = np.random.randint(50, 100, size=(30, 5))  # Generate random scores between 50 and 100

# Create a DataFrame with the data
data = {
    'Math': scores[:, 0],
    'English': scores[:, 1],
    'Science': scores[:, 2],
    'History': scores[:, 3],
    'Geography': scores[:, 4]
}

df = pd.DataFrame(data)

# Create a correlation matrix
corr_matrix = df.corr()

# Set up subplots
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Plot 1: Scatter Plot
axes[0].scatter(df['Math'], df['English'])
axes[0].set_xlabel('Math Scores')
axes[0].set_ylabel('English Scores')
axes[0].set_title('Scatter Plot')

# Plot 2: Bubble Chart
bubble_sizes = np.random.randint(30, 200, size=len(df))  # Generate random sizes for bubbles
scatter = axes[1].scatter(df['Math'], df['English'], s=bubble_sizes, alpha=0.7, cmap='viridis')
plt.colorbar(scatter, ax=axes[1])
axes[1].set_xlabel('Math Scores')
axes[1].set_ylabel('English Scores')
axes[1].set_title('Bubble Chart')

# Plot 3: Heatmap
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', ax=axes[2])
axes[2].set_title('Correlation Heatmap')

# Adjust layout
plt.tight_layout()

# Show plots
plt.show()
````

## 4. **Visualizing Proportions**

When you need to represent parts of a whole, consider:

- **Pie Chart**: Suitable for displaying the contribution of each category to a whole.

- **Stacked Bar Chart**: Shows the composition of different categories as parts of a whole.



Certainly! Let's go through examples and explanations for visualizing proportions using Pie Charts and Stacked Bar Charts:

#### a. Pie Chart

**Example:**

Suppose you have a dataset representing the distribution of expenses in a household budget.

```python
import matplotlib.pyplot as plt

categories = ['Housing', 'Food', 'Transportation', 'Entertainment', 'Other']
expenses = [1500, 800, 600, 300, 400]

plt.pie(expenses, labels=categories, autopct='%1.1f%%', startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Household Expenses')
plt.show()
```

**Explanation:**

- A **pie chart** is used to represent the parts of a whole. Each "slice" of the pie represents a category, and the size of each slice is proportional to the value it represents.
- `expenses` contains the values for each category, and `categories` contains the labels for each slice.
- `autopct='%1.1f%%'` adds the percentage labels to each slice.
- `startangle=140` sets the starting angle of the pie chart. This is useful for rotating the chart to make it more visually appealing.
- `plt.axis('equal')` ensures that the pie chart is drawn as a circle.

In the resulting pie chart, you can easily see the proportion of each expense category in the household budget.

#### b. Stacked Bar Chart

**Example:**

Suppose you have a dataset representing the sales distribution of different products across different regions.

```python
import matplotlib.pyplot as plt

products = ['Product A', 'Product B', 'Product C']
regions = ['North', 'South', 'East', 'West']
sales = [
    [100, 200, 150, 120],  # Product A
    [150, 120, 180, 140],  # Product B
    [200, 160, 100, 120]   # Product C
]

fig, ax = plt.subplots()

bar_width = 0.3
index = range(len(regions))

for i, product in enumerate(products):
    plt.bar(
        [p + i * bar_width for p in index], 
        sales[i], 
        width=bar_width, 
        label=f'{product}'
    )

plt.xlabel('Regions')
plt.ylabel('Sales')
plt.title('Product Sales by Region')
plt.xticks([p + bar_width for p in index], regions)
plt.legend()
plt.show()
```

**Explanation:**

- A **stacked bar chart** is used to show the composition of different categories as parts of a whole.
- In this example, `products` contains the different products, `regions` contains the different regions, and `sales` is a 2D list containing sales data for each product in each region.
- We use a loop to iterate through each product and create a set of bars for each region.
- The bars for each product are stacked on top of each other, showing the total sales for each product in each region.
- `plt.legend()` adds a legend to the chart to distinguish between the products.

In the resulting stacked bar chart, you can easily compare the sales distribution of different products across various regions.

These visualizations are helpful when you need to represent parts of a whole, whether it's expenses in a budget or sales distribution across different categories and regions.

````python
import matplotlib.pyplot as plt

# Data for Pie Chart
categories = ['Housing', 'Food', 'Transportation', 'Entertainment', 'Other']
expenses = [1500, 800, 600, 300, 400]

# Data for Stacked Bar Chart
products = ['Product A', 'Product B', 'Product C']
regions = ['North', 'South', 'East', 'West']
sales = [
    [100, 200, 150, 120],  # Product A
    [150, 120, 180, 140],  # Product B
    [200, 160, 100, 120]   # Product C
]

# Create subplots
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Plot 1: Pie Chart
axes[0].pie(expenses, labels=categories, autopct='%1.1f%%', startangle=0)
axes[0].axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
axes[0].set_title('Household Expenses')

# Plot 2: Stacked Bar Chart
bar_width = 0.2
index = range(len(regions))

for i, product in enumerate(products):
    axes[1].bar(
        [p + i * bar_width for p in index], 
        sales[i], 
        width=bar_width, 
        label=f'{product}'
    )

axes[1].set_xlabel('Regions')
axes[1].set_ylabel('Sales')
axes[1].set_title('Product Sales by Region')
axes[1].set_xticks([p + bar_width for p in index])
axes[1].set_xticklabels(regions)
axes[1].legend()

# Adjust layout
plt.tight_layout()

# Show plots
plt.show()
````

## 5. **Geographical Data**

When you have data associated with geographical locations:

- **Choropleth Map**: Represents data values on a map, typically using color intensity.

- **Bubble Map**: Similar to a choropleth map, but with bubbles of different sizes representing data values.



Certainly! Let's go through examples and explanations for visualizing geographical data using Choropleth Maps and Bubble Maps:

#### a. Choropleth Map

**Example:**

Suppose you have a dataset containing the population density of different states in the United States.

```python
import geopandas as gpd
import matplotlib.pyplot as plt

# Load a map of the United States
us_map = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))

# Sample data for population density
population_density = {
    'California': 250,
    'Texas': 120,
    'New York': 350,
    'Florida': 180,
    'Illinois': 280
}

# Merge data with the map
us_map['Population Density'] = us_map['name'].map(population_density)

# Plot the choropleth map
us_map.plot(column='Population Density', cmap='OrRd', legend=True, legend_kwds={'label': "Population Density per sq. km"})
plt.title('Population Density in US States')
plt.show()
```

**Explanation:**

- A **choropleth map** is used to represent data values on a map, typically using color intensity. It's a way to visualize data for different regions.
- In this example, `geopandas` is used to handle geographical data. We load a map of the United States and assign it to `us_map`.
- `population_density` is a dictionary containing sample data for population density of different states.
- We merge the data with the map using `us_map['Population Density'] = us_map['name'].map(population_density)`.
- The `plot` function is used to create the choropleth map. The `column` parameter specifies the column to use for coloring the map, and `cmap` sets the color map.

In the resulting choropleth map, you can see the population density of different U.S. states represented using color intensity.

#### b. Bubble Map

**Example:**

Suppose you have data on cities and their populations.

```python
import geopandas as gpd
import matplotlib.pyplot as plt

# Load a map of the world
world_map = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))

# Sample data for cities and populations
cities = {
    'New York': (40.7128, -74.0060),  # Latitude and Longitude of New York
    'Los Angeles': (34.0522, -118.2437),
    'Chicago': (41.8781, -87.6298),
    'Houston': (29.7604, -95.3698),
    'Phoenix': (33.4484, -112.0740)
}

city_populations = {
    'New York': 8500000,
    'Los Angeles': 3980400,
    'Chicago': 2716000,
    'Houston': 2323000,
    'Phoenix': 1682000
}

# Create a GeoDataFrame for the cities
cities_df = gpd.GeoDataFrame(geometry=gpd.points_from_xy([lng for lat, lng in cities.values()], [lat for lat, lng in cities.values()]))
cities_df['City'] = list(cities.keys())
cities_df['Population'] = list(city_populations.values())

# Plot the bubble map
world_map.boundary.plot()
cities_df.plot(marker='o', color='red', markersize=cities_df['Population'] / 200000, ax=plt.gca())
plt.title('Population Distribution in Cities')
plt.show()
```

**Explanation:**

- A **bubble map** is similar to a choropleth map, but instead of coloring regions, it places bubbles at specific geographic coordinates.
- In this example, we use `geopandas` to load a map of the world and assign it to `world_map`.
- `cities` is a dictionary containing sample data for cities and their coordinates (latitude and longitude).
- `city_populations` contains sample data for the populations of these cities.
- We create a GeoDataFrame for the cities using their coordinates.
- The `plot` functions are used to plot the boundaries of the world map and the cities as red bubbles. The size of the bubbles is determined by the population.

In the resulting bubble map, you can see bubbles representing cities with sizes proportional to their populations. This allows for a visual representation of population distribution across different cities.

Both choropleth maps and bubble maps are powerful tools for visualizing geographical data, allowing for the representation of data in a spatial context.

````python
import geopandas as gpd
import matplotlib.pyplot as plt

# Choropleth Map
us_map = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
population_density = {
    'United States': 35,  # Sample data for demonstration
    'Canada': 4,
    'Mexico': 10,
    'Greenland': 0.1
}
us_map['Population Density'] = us_map['name'].map(population_density)

# Bubble Map
world_map = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
cities = {
    'New York': (40.7128, -74.0060),
    'Los Angeles': (34.0522, -118.2437),
    'Chicago': (41.8781, -87.6298),
    'Houston': (29.7604, -95.3698),
    'Phoenix': (33.4484, -112.0740)
}
city_populations = {
    'New York': 8500000,
    'Los Angeles': 3980400,
    'Chicago': 2716000,
    'Houston': 2323000,
    'Phoenix': 1682000
}
cities_df = gpd.GeoDataFrame(geometry=gpd.points_from_xy([lng for lat, lng in cities.values()], [lat for lat, lng in cities.values()]))
cities_df['City'] = list(cities.keys())
cities_df['Population'] = list(city_populations.values())

# Set up subplots
fig, axes = plt.subplots(1, 2, figsize=(15, 5))

# Plot Choropleth Map
us_map.plot(column='Population Density', cmap='OrRd', legend=True, legend_kwds={'label': "Population Density per sq. km"}, ax=axes[0])
axes[0].set_title('Population Density in North America')

# Plot Bubble Map
world_map.boundary.plot(ax=axes[1])
cities_df.plot(marker='o', color='red', markersize=cities_df['Population'] / 200000, ax=axes[1])
axes[1].set_title('Population Distribution in Cities')

# Adjust layout
plt.tight_layout()

# Show plots
plt.show()
````

## 6. **Hierarchical Data**

When dealing with data with a hierarchical structure:

- **Tree Map**: Displays hierarchical data in a nested, rectangular layout.




Certainly! Let's go through an example and explanation for visualizing hierarchical data using a Tree Map:

### Tree Map

**Example:**

Suppose you have a dataset representing the hierarchical structure of a company's organizational chart.

```python
import matplotlib.pyplot as plt
import squarify

# Sample hierarchical data
data = {
    'CEO': 100,
    'Manager 1': 60,
    'Manager 2': 40,
    'Employee 1': 20,
    'Employee 2': 30,
    'Employee 3': 10
}

# Create a Tree Map
plt.figure(figsize=(8, 6))
squarify.plot(sizes=data.values(), label=data.keys(), color=["red", "green", "blue", "purple", "orange", "yellow"], alpha=0.6)
plt.title('Organizational Chart')
plt.axis('off')
plt.show()
```

**Explanation:**

- A **tree map** is used to display hierarchical data in a nested, rectangular layout. Each rectangle represents a category, and its size is proportional to a value (e.g., size, weight, etc.).
- In this example, `data` is a dictionary representing the hierarchical structure of the organizational chart. The keys represent positions, and the values represent some measure (e.g., number of employees, revenue, etc.).
- We use the `squarify` library to create the tree map. `sizes` contains the values, and `label` contains the labels for each category.
- `color` parameter specifies the colors of the rectangles. Each category is assigned a different color.
- `alpha=0.6` sets the transparency of the rectangles.

In the resulting tree map, you can see the hierarchical structure of the organization, with rectangles representing different positions. The size of each rectangle is proportional to the value it represents (in this case, an arbitrary measure).

Tree maps are useful for visualizing hierarchical data, allowing you to quickly grasp the relative sizes and proportions of different categories within a hierarchy.


````python

import matplotlib.pyplot as plt
import squarify

# Sample hierarchical data
data = {
    'CEO': 100,
    'Manager 1': 60,
    'Manager 2': 40,.
    'Employee 1': 20,
    'Employee 2': 30,
    'Employee 3': 10
}

# Create a Tree Map
plt.figure(figsize=(8, 6))
squarify.plot(sizes=data.values(), label=data.keys(), color=["red", "green", "blue", "purple", "orange", "yellow"], alpha=0.6)
plt.title('Organizational Chart')
plt.axis('off')
plt.show()
````

## 7. **Time Series Data**

For data that varies over time:

- **Line Chart**: Particularly effective for showing trends and patterns in time series data.

- **Area Chart**: Similar to a line chart but with the area under the line filled, which can help emphasize the cumulative effect.



Certainly! Let's go through examples and explanations for visualizing time series data using Line Charts and Area Charts:

#### a. Line Chart

**Example:**

Suppose you have a dataset representing monthly sales over a year.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample time series data
```

## 8. **Network Data**

When visualizing relationships between nodes:

- **Network Graph**: Useful for displaying connections and relationships in a network.


Creating a network graph requires a specific set of libraries, such as NetworkX for creating and analyzing the structure of the graph, and Matplotlib for visualization. Here's an example of how you can create a simple network graph:

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a simple network graph
G = nx.Graph()

# Add nodes
nodes = ['A', 'B', 'C', 'D', 'E']
G.add_nodes_from(nodes)

# Add edges
edges = [('A', 'B'), ('B', 'C'), ('C', 'D'), ('D', 'E'), ('E', 'A')]
G.add_edges_from(edges)

# Draw the network graph
pos = nx.spring_layout(G)  # Positioning of nodes
nx.draw_networkx_nodes(G, pos, node_color='skyblue', node_size=1000)
nx.draw_networkx_edges(G, pos, edge_color='gray', width=1)
nx.draw_networkx_labels(G, pos, font_size=12, font_color='black')

# Display the graph
plt.title('Simple Network Graph')
plt.axis('off')
plt.show()
```

Explanation:

- The code creates a simple undirected graph with five nodes (A, B, C, D, E) and five edges connecting them.
- `nx.Graph()` creates an empty graph.
- `G.add_nodes_from(nodes)` adds nodes to the graph.
- `G.add_edges_from(edges)` adds edges between nodes.
- `pos = nx.spring_layout(G)` calculates the positions of nodes for visualization.
- `nx.draw_networkx_nodes`, `nx.draw_networkx_edges`, and `nx.draw_networkx_labels` are used to draw the nodes, edges, and labels respectively.

This code will display a simple network graph. You can customize the nodes, edges, and their properties to represent different types of networks.

````python
import networkx as nx
import matplotlib.pyplot as plt

# Create a simple network graph
G = nx.Graph()

# Add nodes
nodes = ['A', 'B', 'C', 'D', 'E']
G.add_nodes_from(nodes)

# Add edges
edges = [('A', 'B'), ('B', 'C'), ('C', 'D'), ('D', 'E'), ('E', 'A')]
G.add_edges_from(edges)

# Draw the network graph
pos = nx.spring_layout(G)  # Positioning of nodes
nx.draw_networkx_nodes(G, pos, node_color='skyblue', node_size=1000)
nx.draw_networkx_edges(G, pos, edge_color='gray', width=1)
nx.draw_networkx_labels(G, pos, font_size=12, font_color='black')

# Display the graph
plt.title('Simple Network Graph')
plt.axis('off')
plt.show()
````


## 9. **Comparing Multiple Variables**

When you want to compare multiple variables simultaneously:

- **Radar Chart**: Shows multivariate data in the form of a two-dimensional chart.

Creating a radar chart requires the use of the `matplotlib` library. Here's an example of how you can create a radar chart to compare multiple variables:

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the variables and their values
categories = ['Category A', 'Category B', 'Category C', 'Category D', 'Category E']
values = [4, 2, 5, 1, 3]

# Number of categories
num_vars = len(categories)

# Calculate angles for each category
angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()

# Make plot circular
values += values[:1]
angles += angles[:1]

# Create the radar chart
plt.figure(figsize=(8, 8))
ax = plt.subplot(111, polar=True)

# Plot the data
plt.plot(angles, values, color='b', linewidth=2)

# Fill area under the line
plt.fill(angles, values, color='b', alpha=0.25)

# Add labels and title
plt.xticks(angles[:-1], categories)
plt.yticks([1, 2, 3, 4, 5], color='grey', size=10)
plt.ylim(0, 5)
plt.title('Radar Chart Example', size=20, color='b')

# Show the plot
plt.show()
```

Explanation:

- The code creates a radar chart to compare multiple variables (in this case, five categories).
- `categories` represents the variables, and `values` represents their corresponding values.
- The code calculates the angles for each category to position them evenly around the circle.
- The plot is created using `plt.subplot(111, polar=True)` to set up a polar coordinate system.
- The data is plotted using `plt.plot` and the area under the line is filled with `plt.fill`.
- Labels, ticks, and title are added to the chart for clarity.

This code will display a radar chart comparing the values of different categories. You can customize the categories and their values to represent different variables.

````python
import matplotlib.pyplot as plt
import numpy as np

# Define the variables and their values
categories = ['Category A', 'Category B', 'Category C', 'Category D', 'Category E']
values = [4, 2, 5, 1, 3]

# Number of categories
num_vars = len(categories)

# Calculate angles for each category
angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()

# Make plot circular
values += values[:1]
angles += angles[:1]

# Create the radar chart
plt.figure(figsize=(8, 8))
ax = plt.subplot(111, polar=True)

# Plot the data
plt.plot(angles, values, color='b', linewidth=2)

# Fill area under the line
plt.fill(angles, values, color='b', alpha=0.25)

# Add labels and title
plt.xticks(angles[:-1], categories)
plt.yticks([1, 2, 3, 4, 5], color='grey', size=10)
plt.ylim(0, 5)
plt.title('Radar Chart Example', size=20, color='b')

# Show the plot
plt.show()
````
